```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("old_put_all.csv")
df = df.iloc[0:10]
df
```

```python
plt.rcParams['axes.facecolor'] = '#F9F5F0'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 12
plt.rcParams['font.weight'] = 'bold'
```

# Put All

```python
plt.figure(figsize=(10,6))
plt.plot(df["Clients"], df["Throughput(req/s)"], color='#ED3F27', marker='o', linewidth=1.5)
plt.grid(True)
plt.xticks(df["Clients"])
plt.annotate("Saturation Reached", 
             xy=(13, 850),
             xytext=(13, 750),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1),
             va='center',
             ha='center')
plt.title("Put All Workload\n Throughput vs Number of Concurrent Clients")
plt.ylabel("Throughput(Req/sec)")
plt.xlabel("Number of Concurrent Clients")
plt.savefig("old_put_all_thr_vs_clients.jpeg", dpi = 1200)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Clients"], df["AvgRespTime(ms)"], color='#ED3F27', marker='o', linewidth=1.5)
plt.grid(True)
plt.xticks(df["Clients"])
plt.title("Put All Workload\nLatency vs Number of Concurrent Clients")
plt.ylabel("Latency(ms)")
plt.xlabel("Number of Concurrent Clients")
plt.savefig("old_put_all_lat_vs_clients.jpeg", dpi = 1200)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Clients"], df["ServerCPU(%)"], color='#ED3F27', marker='o', linewidth=1.5)
plt.grid(True)
plt.xticks(df["Clients"])
plt.title("Put All Workload\nCPU Utilization vs Number of Concurrent Clients")
plt.ylabel("CPU Utilization %")
plt.xlabel("Number of Concurrent Clients")
plt.savefig("old_put_all_cpu_vs_clients.jpeg", dpi = 1200)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Clients"], df["MaxDiskIO(%)"], color='#ED3F27', marker='o', linewidth=1.5)
plt.grid(True)
plt.xticks(df["Clients"])
plt.title("Put All Workload\nIO Utilization vs Number of Concurrent Clients")
plt.ylabel("IO Utilization %")
plt.xlabel("Number of Concurrent Clients")
plt.savefig("old_put_all_io_vs_clients.jpeg", dpi = 1200)
plt.show()
```
